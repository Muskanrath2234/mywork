# imputation techqiues

## 1st - mean value imputation
          Mean imputation is often used when the missing values are numerical and the distribution 
          of the variable is approximately normal.

#### this is work well with normally distributed data

```python
sns.displot(df['age'])
```

```python
df.age.isnull().sum()
```

```python
df['age_mean']= df['age'].fillna(df['age'].mean())
```

```python
df[['age_mean','age']]
```

### 2nd- median value imputation
       Median imputation is preferred when the distribution is skewed, as the median is less        
       effected outliers than the mean.

```python
df['age_median']= df['age'].fillna(df['age'].median())
```

```python
df[['age_mean','age_median','age']]
```

```python
df.shape
```

### Mode imputation 
   mode imputation is suitable for categorical variables or 
   numerical variables with a small number of unique values.

```python
df.head()
```

```python
df['embarked'].isnull().sum()
```

```python
df[df["embarked"].isnull()]
```

```python
# unique value
```

```python
df["embarked"].unique()
```

```python
df["embarked"].isnull().sum()
```

```python
df[df["embarked"].notna()]
```

```python
df[df["embarked"].notna()]["embarked"].mode()
```

```python
mode=df[df["embarked"].notna()]["embarked"].mode()[0]
```

```python
mode
```

```python
df["embarked_mode"]=df["embarked"].fillna(mode)
```

```python
df["embarked_mode"]
```
